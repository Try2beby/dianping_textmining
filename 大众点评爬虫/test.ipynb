```python
import requests


def test_proxy(server, port, test_url="http://www.baidu.com"):
    # 构造代理服务器信息
    proxies = {"http": f"http://{server}:{port}", "https": f"https://{server}:{port}"}

    try:
        # 发起请求
        response = requests.get(test_url, proxies=proxies, timeout=5)  # 设置超时为5秒
        # 判断响应状态码
        if response.status_code == 200:
            print(
                f"代理 {server}:{port} 可用，响应时间：{response.elapsed.total_seconds()}秒"
            )
        else:
            print(f"代理 {server}:{port} 不可用，状态码：{response.status_code}")
    except requests.exceptions.ProxyError:
        print(f"代理 {server}:{port} 不可用，代理错误")
    except requests.exceptions.ConnectTimeout:
        print(f"代理 {server}:{port} 连接超时")
    except requests.exceptions.Timeout:
        print(f"代理 {server}:{port} 请求超时")
    except requests.exceptions.RequestException as e:
        print(f"代理 {server}:{port} 请求失败：{e}")


# 示例：测试代理服务器
# test_proxy("117.93.180.231", "9000")
# test_proxy("localhost", "7890")
test_proxy("117.92.97.127", "22641")
```
